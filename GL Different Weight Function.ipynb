```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import random
from GraphLaplacian.binary import GraphLaplacian

sns.set()
```

```python
N = 1000
L = 80

X0 = [[0, 10*i/(L/4)] for i in range(np.int32(L/4))]
X0 += [[5*i/(L/8), 0] for i in range(np.int32(L/8))]
X0 += [[5*i/(L/8), 10] for i in range(np.int32(L/8))]

X0 += [[10, 10*i/(L/4)] for i in range(np.int32(L/4))]
X0 += [[5 + 5*i/(L/8), 0] for i in range(np.int32(L/8))]
X0 += [[5 + 5*i/(L/8), 10] for i in range(np.int32(L/8))]

Y0 = [0 for i in range(np.int32(L/2))]
Y0 += [1 for i in range(np.int32(L/2))]
Y0 = np.array(Y0)

## Generate X Data
X1 = [[random.random()*10, random.random()*10] for i in range(np.int32(N))]

len(X0)
```

```python
GL = GraphLaplacian(X0, Y0, X1)
GL.plot_problem()

# def weight_func(xi, xj, sigm):
#     np.sqrt(np.dot(xi - xj,xi - xj))
#     return 1 / (1 + np.sqrt(np.dot(xi - xj,xi - xj)))

# GL.weight_function = weight_func
```

```python
# A,b = GL.create_linear_system()
# U = np.linalg.solve(A,b)
def plots(X0, X1, U, Y):
    X = np.array(X0 + X1)
    L = len(X0)
    plt.figure(figsize=(15,5))

    ax2 = plt.subplot(121)
    # Create an array with the colors you want to use
    colors = ["#ff6361", "#003f5c"]
    # Set your custom color palette
    sns.set_palette(sns.color_palette(colors))
    df = pd.DataFrame(X, columns=["x1","x2"])
    df.loc[:, "RESULT"] = Y
    sns.scatterplot(data=df, x="x1", y="x2", hue="RESULT", ax=ax2)
    ax2.set_title("RESULT")

    ax3 = plt.subplot(122)
    # Create an array with the colors you want to use
    colors = ["#ff6361", "#003f5c"]
    # Set your custom color palette
    sns.set_palette(sns.color_palette(colors))
    df = pd.DataFrame(X, columns=["x1","x2"])
    df.loc[:, "U"] = U
    sns.scatterplot(data=df, x="x1", y="x2", hue="U", ax=ax3, legend="brief")
    ax3.set_title("U Value")

    plt.show()
```
